# Authentication

```python
# Make sure you put "tweeter_api.txt" in your working directory.
with open ("tweeter_api.txt", "r") as myfile:
    data=myfile.readlines()
```

```python
customer_key = data[0]
customer_key = customer_key[:-1]

customer_secret = data[1]
customer_secret = customer_secret[:-1]

access_token = data[2]
access_token = access_token[:-1]

access_secret = data[3]
access_secret = access_secret[:-1]
```

```python
auth = tweepy.OAuthHandler(customer_key, customer_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

# The Crawler (Specific Users Tweets)

```python
# Do not run this code twice or you will lose all the data.

df = pd.DataFrame({'Date':[],
                   "ID":[],
                   "Name":[],
                   "Text":[],
                   "Hashtags":[],
                   "User Mentions":[],
                   "Symbols":[],
                   "Media":[],
                   "Source":[],
                   "Location":[],
                   "Follower count": [],
                   "Friends count":[],
                   "Favorite count":[],
                   "Retweet count":[],
                   "Status count":[]})  
```

```python
def get_tweet(User_name):
    tweets = api.user_timeline(screen_name = User_name,count=500)

    date = []
    ID = []
    name = []
    text = []
    hashtags = []
    user_mentions = []
    symbols = []
    media = []
    source = []
    location = []
    follower_count = []
    friends_count = []
    favorite_count = []
    retweet_count = []
    status_count = []

    for t in tweets:
        if t.created_at is not None:
            date.append(t.created_at)
        else: 
            date.append(np.nan)

        if t.id_str is not None:
            ID.append(t.id_str)
        else:
            t.append(np.nan)

        if t.user.name is not None:
            name.append(t.user.name)
        else:
            name.append(np.nan)

        if t.text is not None:
            text.append(t.text)
        else:
            text.append(np.nan)

        if t.entities['hashtags'] != []:
            tmp = []
            for h in t.entities['hashtags']:
                tmp.append(h['text'])
            hashtags.append(','.join(tmp))
        else:
            hashtags.append(np.nan)

        if t.entities['user_mentions'] != []:
            tmp = []
            for u in t.entities['user_mentions']:
                tmp.append(u['name'])
            user_mentions.append(','.join(tmp))
        else:
            user_mentions.append(np.nan)

        if t.entities['symbols'] != []:
            tmp = []
            for s in t.entities['symbols']:
                tmp.append(s['text'])
            symbols.append(','.join(tmp))
        else:
            symbols.append(np.nan)

        try:
            tmp = []
            for m in t.entities['media']:
                tmp.append(m['media_url'])
            media.append(','.join(tmp))
        except:
            media.append(np.nan)
        
        if t.source is not None:
            source.append(t.source)
        else:
            source.append(np.nan)

        if t.user.location is not None:
            location.append(t.user.location)
        else:
            source.append(np.nan)

        if t.user.followers_count is not None:
            follower_count.append(t.user.followers_count)
        else:
            follower_count.append(np.nan)

        if t.user.friends_count is not None:
            friends_count.append(t.user.friends_count)
        else:
            friends_count.append(np.nan)

        if t.favorite_count is not None:
            favorite_count.append(t.favorite_count)
        else:
            favorite_count.append(np.nan)

        if t.retweet_count is not None:
            retweet_count.append(t.retweet_count)
        else:
            retweet_count.append(np.nan)

        if t.user.statuses_count is not None:
            status_count.append(t.user.statuses_count)
        else:
            status_count.append(np.nan)


    data = pd.DataFrame({'Date':date,
                       "ID":ID,
                       "Name":name,
                       "Text":text,
                       "Hashtags":hashtags,
                       "User Mentions":user_mentions,
                       "Symbols":symbols,
                       "Media": media,
                       "Source":source,
                       "Location":location,
                       "Follower count": follower_count,
                       "Friends count":friends_count,
                       "Favorite count":favorite_count,
                       "Retweet count":retweet_count,
                       "Status count":status_count})
    return data
```

```python
list_of_account = [['YouTube', 'Twitter', 'instagram', 'Google', 'Facebook', 'netflix',
                  'Microsoft', 'Tesla', 'Apple', 'amazon'],
                   ['CHANEL', 'Starbucks', 'VictoriasSecret', 'marcjacobs',
                  'hm', 'Burberry','Dior', 'nikestore', 'LouisVuitton', 'gucci', 'dolcegabbana', 'Versace',
                  'WholeFoods', 'McDonalds', 'Wendys', 'CocaCola'],
                   ['cnnbrk', 'CNN', 'nytimes',
                  'BBCBreaking', 'SportsCenter', 'espn', 'BBCWorld', 'TheEconomist', 'Reuters',
                  'WSJ'], 
                   ['BarackObama', 'POTUS45', 'HillaryClinton', 'JoeBiden',  'Pontifex', 'KamalaHarris'],
                   ['elonmusk', 'JeffBullas', 'SMExaminer', 'KimReynolds', 'JayBaer', 'KimGarst','PamMktgNut',
                   'PeterShankman', 'PegFitzpatrick', 'MarshaCollier', 'NealShaffer', 'BrianCarter', 'RebekahRadice',
                   'JoePulizzi', 'MarkWSchaefer', 'GaryVee', 'TedRubin', 'GuyKawasaki', 'BryanKramer','JasonFalls',
                   'ChrisBrogan', 'MitchJoel'],
                   ['justinbieber', 'katyperry', 'rihanna', 'taylorswift13', 'ladygaga', 'ArianaGrande','selenagomez',
                   'jtimberlake', 'britneyspears', 'ddlovato'], 
                   ['TheEllenShow', 'jimmyfallon', 'akshaykumar', 'KevinHart4real', 'TheRock', 'RobertDowneyJr',
                   'vindiesel', 'TomCruise', 'ChrisEvans', 'GalGadot']] # Insert your list of accounts


industries = ['Tech Company', 'Non-tech Company', 'News', 'Politician', 'Influencer', 'Singer','Actor']

for acc,ind in zip(list_of_account, industries):
    for a in acc:
        tmp = get_tweet(a)
        tmp['Industries'] = ind
        df = df.append(tmp)
```
